### 1. Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

1. Pyplot (State-based) Interface:

This is similar to MATLAB-style plotting.

It uses functions from the matplotlib.pyplot module to create and manipulate plots.

2. Object-Oriented (OO) Interface:

This is a more explicit and flexible approach.

You create figure and axes objects and then call methods on them.

The Object-Oriented (OO) interface is the recommended approach, especially for creating complex plots, reusing plot components, creating multiple subplots or fine-tuning layouts or maintaining clarity and modularity in larger codebases.

### 4. Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Data
x = np.linspace(-4, 4, 100)
y = x**2

# Plot using pyplot approach
plt.figure()
plt.plot(x, y)
plt.title("Plot of y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

### 5. Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Seed and data generation
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot
ax1.scatter(random_data_x, random_data_y, alpha=0.5)
ax1.plot(x, y, color='red', label='y = x^2')
ax1.set_title('Scatter Plot with Curve')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.legend()

# Bar plot
ax2.bar(names, values, color='skyblue')
ax2.set_title('Fruit Quantity Bar Plot')
ax2.set_ylabel('Quantity')

# Layout adjustment
plt.tight_layout()
plt.show()
```
