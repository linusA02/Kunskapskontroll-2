```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Setting styles for plots
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Load the dataset
data = pd.read_csv(r"C:\Users\epicp\Downloads\housing.csv")
```

```python
# Basic dataset information
print("First 5 rows of the dataset:")
print(data.head())

print("\nDataset Info:")
print(data.info())

print("\nSummary Statistics:")
print(data.describe())

# Checking for missing values
print("\nMissing values:")
print(data.isnull().sum())
```

```python
# Distribution of numerical features
numerical_features = data.select_dtypes(include=[np.number]).columns.tolist()
data[numerical_features].hist(bins=30, figsize=(15, 10), color='skyblue')
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
# Scatter plots of target variable vs some features
target_col = 'median_house_value'
features_to_plot = [col for col in numerical_features if col != target_col]

for feature in features_to_plot:
    plt.figure()
    sns.scatterplot(x=data[feature], y=data[target_col])
    plt.title(f"{target_col} vs {feature}")
    plt.xlabel(feature)
    plt.ylabel(target_col)
    plt.show()

# Boxplots for categorical features (if any exist)
categorical_features = data.select_dtypes(include=["object"]).columns.tolist()
for feature in categorical_features:
    plt.figure()
    sns.boxplot(x=data[feature], y=data[target_col])
    plt.title(f"{target_col} by {feature}")
    plt.xlabel(feature)
    plt.ylabel(target_col)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Value counts of categorical features
for feature in categorical_features:
    print(f"\nValue counts for {feature}:")
    print(data[feature].value_counts())
```
